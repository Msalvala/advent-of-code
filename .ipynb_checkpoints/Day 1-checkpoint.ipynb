```python
import os

data = open(os.getcwd() + '/input.txt')
report = []

for line in data:
    
    report.append(int(line.replace('\n', '')))
    
print(report)
```

```python
def sep(num_list,position,position_2):
    
    pos_num = num_list[position]
    rem_num = num_list[:position]+num_list[position+1:]
    rem_num_2 = rem_num[:position_2]+num_list[position_2+1:]
    
    for i in range(len(rem_num)):  
        
        for j in range(len(rem_num_2)):
        
            pot_ans = pos_num+rem_num[i]+rem_num_2[j]
        
            if pot_ans==2020:
        
                print(str(pos_num)+" times "+str(rem_num[i])+" times "+str(rem_num_2[j])+" equals "+str(rem_num[i]*pos_num*rem_num_2[j])+" !!!!!!!")
            
for i in range(len(report)):
    
    for j in range(len(report)-1):
        
        sep(num_list=report,position=i,position_2=j)
        


```
